## Performance Test - Sum of multiples

The function below times all functions in perf_test.py, which contains three different implementations of the sum of multiples exercise. It shows a chart comparing the different functions with increasing input numbers.


```python
import numpy as np
import timeit
from matplotlib import pyplot as plt
import ast
from perf_test import *


def get_functions(file):
    source = open(file).read()
    functions = [f.name for f in ast.parse(source).body
             if isinstance(f, ast.FunctionDef)]
    return functions

def measure(func, limit):
    start_time = timeit.default_timer()
    r = range(1, limit)
    dt = []
    for i in r:
        eval(func)(i, [3, 5])
        dt.append(timeit.default_timer() - start_time)
    plt.plot(r, dt, label=func)
    plt.legend(loc="upper left")
    

file = 'perf_test.py'    
limit = 10000

for func in get_functions('perf_test.py'):
    measure(func, limit)
```
